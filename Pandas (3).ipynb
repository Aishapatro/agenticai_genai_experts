# Importing multiple CSV files in DataFrame 

```python
import pandas as pd
import glob

# Folder path do, CSV file ka nahi
path = r'
filenames = glob.glob(path + "/*.csv")
print(filenames)

covid = pd.DataFrame()

for f in filenames:
    df = pd.read_csv(f)
    covid = pd.concat([covid, df], ignore_index=True, sort=True)

print(covid.shape)
```

```python
import pandas as pd
import glob

path = r'C:\\Users\\Aisharani Patro\\OneDrive\\Desktop\\data.csv'

filenames = glob.glob(path + "/*.csv")

print(filenames)

dfs = []

for f in filenames:
    try:
        df = pd.read_csv(f)
        dfs.append(df)
    except Exception as e:
        print("Error in file:", f)
        print(e)

covid = pd.concat(dfs, ignore_index=True)

print(covid.head())
```
