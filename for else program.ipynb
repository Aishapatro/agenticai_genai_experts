```python
# check the number factorial or not
num=int(input("Enter a number:"))

fact=1

for i in range(1,num+1):
    fact=fact*i

    if fact==num:
         print(num,"is a factorial of",i)
         break
    if fact>num:
        print(num,"is not a factorial number")
        break
else:
    print(num,"is not a factorial number")
```

```python
# check palindrome or not
text=input("Enter a string or number:")

n=len(text)

for i in range(n//2):
     if text[i]!=text[n-i-1]:
         print("Not a palindrome")
         break
else:
    print("palindrome")
```

```python
# check palindrome or not

text=input("Enter a string or number:")

left=0
right=len(text)-1

while left<right:
    if text[left]!=text[right]:
        print("Not a palindrome")
        break
    left+=1
    right-=1
else:
    print("palindrome")
```
