```python
import numpy as np
import pandas as pd
```

```python
Country = ['India','USA','Brazil','Russia','UK']
Cases = [45000000,103000000,37000000,22000000,24000000]
Deaths = [530000,1100000,700000,390000,220000]
Recovered = [44400000,100000000,36000000,21500000,23500000]
```

```python
Total_Cases = np.sum(Cases)
print(Total_Cases)
```

```python
Average_Cases = np.mean(Cases)
print(Average_Cases)
```

```python
Highest_Cases = np.max(Cases)
print(Highest_Cases)
```

```python
Lowest_Deaths = np.min(Deaths)
print(Lowest_Deaths)
```

```python
Recovery_Rate = np.array(Recovered) / np.array(Cases) * 100
print(Recovery_Rate)
```

```python
import matplotlib.pyplot as plt
```

# Bar Graph 

```python
Country = ['India','USA','Brazil','Russia','UK']
Cases = [45000000,103000000,37000000,22000000,24000000]
plt.bar(Country, Cases)
plt.title("COVID-19 Total Cases")
plt.xlabel("Country")
plt.ylabel("Cases")
plt.show()
```

# Pie Chart 

```python
Deaths = [530000,1100000,700000,390000,220000]
plt.pie(Deaths, labels=Country, autopct='%1.1f%%')
plt.title("COVID-19 Death Distribution")
plt.show()
```

# Line Graph 

```python
Recovered = [44400000,100000000,36000000,21500000,23500000]
plt.plot(Country, Recovered)
plt.title("COVID-19 Recovery Cases")
plt.xlabel("Country")
plt.ylabel("Recovered")
plt.show()
```

# Compare Cases and Recoveries 

```python
import numpy as np
x = np.arange(len(Country))
width = 0.35
plt.bar(x - width/2, Cases, width, label='Cases')
plt.bar(x + width/2, Recovered, width, label='Recovered')
plt.xticks(x, Country)
plt.title("Cases vs Recoveries")
plt.legend()
plt.show()
```
