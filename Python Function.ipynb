# Postional Argument 

```python
def person(name,age):
    print(name)
    print(age)

person('nit',22)
```

```python
def person(name,age):
    print(name)
    print(age)

person('nit')
```

```python
def person(name,age):
    print(name)
    print(age)

person(22)
```

```python
def person(name,age):
    print(name)
    print(age)

person(22,'nit')
```

```python
def person(name,age):
    print(name)
    print(age+1)

person(22,'nit')
```

# Keyword Argument 

```python
def person(name,age):
    print(name)
    print(age+1)

person(age=22,name='nit')
```

```python
def person(name,age,mob):
    print(name)
    print(age+1)
    print(mob)

person(23,'nit',345)
```

```python
def person(name,age,mob):
    print(name)
    print(age+1)
    print(mob)

person(age=23,name='nit',345)
```

# Default Argument 

```python
def person(name,age):
    print(name)
    print(age) 
    
person('nit')  
```
