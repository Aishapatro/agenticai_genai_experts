```python
%matplotlib inline 
plt.rcParams['figure.figsize']=6,2
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1=sns.distplot(df["InternetUsers"])
```

```python
vis2=sns.displot(df["InternetUsers"])
```

```python
vis3=sns.distplot(df["InternetUsers"],bins=15)
```

```python
vis4=sns.boxplot(data=df,x="IncomeGroup",y='BirthRate')  #BI-VARIATE ANALYSIS
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate')
```

```python

vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)  #lm-linear model
```

```python

vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True)  #lm-linear model
```

```python

vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup')  #hue parameter for color
```

```python

vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True,hue='IncomeGroup')  #hue parameter for color
```

```python
vis6 = sns.lmplot(data =df,x ='InternetUsers',y='BirthRate',
                  fit_reg=False,hue ='IncomeGroup')
```
