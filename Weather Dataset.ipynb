```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
City = np.array([
    'Delhi',
    'Mumbai',
    'Chennai',
    'Kolkata',
    'Banglore',
    'Hyderabad',
    'Pune'
])

Temperature = np.array([
    42,
    34,
    38,
    29,
    33,
    31,
    30
])

Humidity = np.array([
    40,
    75,
    70,
    65,
    55,
    50,
    60
])

Rainfall = np.array([
    5,
    20,
    18,
    15,
    25,
    8,
    12
])

weather = pd.DataFrame({
    'City': City,
    'Temperature': Temperature,
    'Humidity': Humidity,
    'Rainfall': Rainfall
})

print(weather)
```

```python
print(weather.info())
print(weather.shape)
```

```python
Average_Temperature=np.mean(Temperature)
print("Average Temperature :",Average_Temperature)
```

```python
Highest_Temperature=np.max(Temperature)
Hot_City=City[np.argmax(Temperature)]
print("Highest Temperature :",Highest_Temperature)
print("Hottest City:",Hot_City)
```

```python
Lowest_Temperature=np.min(Temperature)
Cold_City=City[np.argmin(Temperature)]
print("Lowest Temperature :", Lowest_Temperature)
print("Coolest City:",Cold_City)    
```

```python
plt.figure(figsize=(10,5)) 
plt.bar(City,Temperature)
plt.bar(City,Temperature)
plt.title("City Temperature Analysis")
plt.xlabel("Cities")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    City,
    Humidity,
    marker='o'
)
plt.title("Humidity Analysis")
plt.xlabel("Cities")
plt.ylabel("Humidity")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    Rainfall,
    labels=City,
    autopct='%1.1f%%'
)
plt.title("Rainfall Disribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(
    Temperature,
    Humidity
)
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(Temperature)
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
